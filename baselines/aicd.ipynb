```python
import pandas as pd
testing_frame = pd.read_json("/kaggle/input/dactyl-nonadversarial/dactyl_testing_split.json")
```

```python
# Use a pipeline as a high-level helper
from transformers import pipeline
from tqdm import tqdm
pipe = pipeline("text-classification", model="PirateXX/AI-Content-Detector")

def data():
    for text in testing_frame["text"].values:
        yield text
y_pred = list()
for result in tqdm(pipe(data(),batch_size=16, max_length=512,truncation=True, return_all_scores=True), total=len(testing_frame)):
    for output in result:
        if output["label"] == "LABEL_0":
            y_pred.append(output["score"])

```

```python
testing_frame["piratexx_pred"] = y_pred
testing_frame = testing_frame.drop(columns="text")
testing_frame[["id","piratexx_pred","target", "domain","model","category"]].to_csv(f"piratexx-results.csv",index=False)
```
